```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
URL2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/diamonds.csv"

df = pd.read_csv(URL2)
```

```python
df.describe()
```

```python
df.sample(6)
```

```python
df.shape
```

```python
df.plot.scatter(x = "price", y= "carat")
```

```python
target = df["price"]
```

```python
features = df[["carat","depth"]]
```

```python
lr2 = LinearRegression()
```

```python
lr2.fit(features, target)
```

```python
lr2.score(features, target)
```

```python
lr2.predict([[carat = 0.3, depth = 60]])
```
